# 1. Clean Data

### clean 15-19 data

Dataset gathered from open-source Canadian Government website which houses immigration, number of citizenship requests and number of Canadian Citizens
Link: https://open.canada.ca/data/en/dataset/9b34e712-513f-44e9-babf-9df4f7256550

Csv file shows life of New Immigrated Canadian Citizens from 2015-2019.

```python
can1519 = pd.read_excel("1519 Canada immigration.xls",
                        na_values= 0,
                        Header = None,
                        index_col = 0,
                        names = ["Country",2015,2016,2017,2018,2019],
                        skiprows = [0,1,238,239,240,241,242,243,244,245],
                        usecols = [0,13,26,39,52,65]
                       )
can1519
```

```python
#filled NA
can1519.fillna(0, inplace=True)
can1519
```

```python
#save the data name it "newdf1519.csv"
can1519.to_csv ('cleanedimmigration1519.csv', 
                    header=True)
```

### Clean 1980-2013 immigration data
