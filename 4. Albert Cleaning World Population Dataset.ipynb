# Import and Cleaning of the World Population Dataset per country from 1979 - 2018

- This dataset is used to do the per-capita of New Canadian Citizens
- Dataset sourced from worldbank
link: https://data.worldbank.org/indicator/sp.pop.totl

```python
#Function to a listexception ( main list, list of things to skip)

def listexception(columnsinworld, columnstoskip):
    for item in columnsinworld:
        if item not in columnstoskip:
            yield item
```

```python
columnstoskipex = [i for i in range(1,23)]
columnsinworldex = [x for x in range(0,63)]
worldcol = listexception(columnsinworldex,columnstoskipex)
listworldcol = list(worldcol)
```

```python
world = pd.read_csv("world.csv",
                    na_values="0",
                    usecols = listworldcol,
                    header = 0,
                    index_col= 0,
                   )
world
```

```python
world.fillna(0, inplace=True)
```

```python
world.to_csv ('cleanedworldnew.csv', header=True)
```
