# Getting Per- Capita of Canadian Citizens Immigrated

    - Divide the original dataset by total population of the country and get percentage
    

```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
dfcancitizen = pd.read_csv("cleanedcanadaimmigration.csv",
                           index_col = "Country"
                          )
dfcancitizen
```

```python
dfpopulation = pd.read_csv("cleanedworldnew.csv",
                           index_col = "Country"
                          )
dfpopulation
```

```python

tempregion = dfcancitizen["Region"]
tempstatus = dfcancitizen["Status"]
```

```python
#combining both datasets
dfcombined = (dfcancitizen / dfpopulation) * 100
dfcombined
```

```python
# adding back the Region and status
dfcombined["Region"]=tempregion
dfcombined["Status"]=tempstatus
dfcombined
```

```python
#drop NA
dfdropped = dfcombined.dropna(how='all')
dfdropped
```

```python
dfdropped.to_csv("cleanedcombined.csv")
```
