## 2. Merge both datasets 

- with both cleaned 1980-2013 and 2015-2019 datasets, Attempted to merge both into one cohesive dataset from 1980-2019

### Loading of both datasets

```python
df1 = pd.read_csv('cleanedimmigration1519.csv', index_col=0)
df1
```

```python
df2 = pd.read_csv('cleanedimmigration.csv', index_col=0)
df2
```

### Merge both datasets

- Merge through country index

```python
immigration = pd.merge(df2, df1, how='inner', right_index = True, left_index = True)
immigration,
```

### Making 2014 column

- 2014 data was missing. I Emailed the govenment for the missing 2014 dataset, no reply since. <br> 
Because it cant be supplied by Statistics Canada I made new column "2014" as average of column '2013' and '2015'.

```python
#make 2014 average of both 2013 and 2015
immigration ['2014'] = (immigration ['2013'] + immigration ['2015']) / 2
immigration
```

```python
#rearranging the columns
var = immigration['2014']
immigration.drop(labels=['2014'], axis=1,inplace = True)
immigration.insert(36, '2014', var)
immigration
```

```python
immigration.to_csv ('cleanedcanadaimmigration.csv', header=True)
```
